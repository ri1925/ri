```python
name_of_customer = input("ENTER YOUR NAME :  ")
print("")
address = input("ENTER YOUR ADDRESS:  ")
print("")
mobile_number = int(input("ENTER YOUR MOBILE NUMBER:   "))
print("")
item1 = "Mat"
price_of_item1 = 750
discount1 = 76

item2 = "Bag"
price_of_item2 = 1500
discount2 = 80

item3 = "Toy"
price_of_item3 = 900
discount3 = 70

item4 = "Bat"
price_of_item4 = 600
discount4 = 45

item5 = "Hat"
price_of_item5 = 790
discount5 = 30


discount_amount1 = price_of_item1 * discount1 / 100
final_price1 = price_of_item1 - discount_amount1

discount_amount2 = price_of_item2 * discount2 / 100
final_price2 = price_of_item2 - discount_amount2

discount_amount3 = price_of_item3 * discount3 / 100
final_price3 = price_of_item3 - discount_amount3

discount_amount4 = price_of_item4 * discount4 / 100
final_price4 = price_of_item4 - discount_amount4

discount_amount5 = price_of_item5 * discount5 / 100
final_price5 = price_of_item5 - discount_amount5


final_price_after_discount = final_price1 + final_price2 + final_price3 + final_price4 + final_price5
gst = final_price_after_discount * 18 / 100
final_price_after_gst = final_price_after_discount + gst
deliverycharge = 60
exact_price = final_price_after_gst + deliverycharge


print("\n-------------------------------------- CUSTOMER BILL ---------------------------------------------------------\n")
print("")

("NAME OF CUSTOMER                 : ", name_of_customer)
print("")
("ADDRESS OF CUSTOMER              : ", address)
print("")
("MOBILE NUMBER OF CUSTOMER        : ", mobile_number)
print("")


print("ITEM THAT CUSTOMER PURCHASED AND THEIR ACTUAL PRICE :")
print("")

print("ITEM1 : ", item1, "                            ACTUAL PRICE                   :   ", price_of_item1,".rs")
print("ITEM2 : ", item2, "                            ACTUAL PRICE                   :   ", price_of_item2,".rs")
print("ITEM3 : ", item3, "                            ACTUAL PRICE                   :   ", price_of_item3,".rs")
print("ITEM4 : ", item4, "                            ACTUAL PRICE                   :   ", price_of_item4,".rs")
print("ITEM5 : ", item5, "                            ACTUAL PRICE                   :   ", price_of_item5,".rs")
print("")


print("ITEM THAT CUSTOMER PURCHASED AND THE AMOUNT OF DISCOUNT :")
print("")

print("ITEM1 : ", item1, "                            AMOUNT OF DISCOUNT THEY GET IS  :   ", discount_amount1,".rs")
print("ITEM2 : ", item2, "                            AMOUNT OF DISCOUNT THEY GET IS  :   ", discount_amount2,".rs")
print("ITEM3 : ", item3, "                            AMOUNT OF DISCOUNT THEY GET IS  :   ", discount_amount3,".rs")
print("ITEM4 : ", item4, "                            AMOUNT OF DISCOUNT THEY GET IS  :   ", discount_amount4,".rs")
print("ITEM5 : ", item5, "                            AMOUNT OF DISCOUNT THEY GET IS  :   ", discount_amount5,".rs")
print("")

print("FINAL PRICE OF ITEM AFTER DISCOUNT IS :")
print("")

print("ITEM1 : ", item1, "                           PRICE AFTER DISCOUNT             :   ", final_price1,".rs")
print("ITEM2 : ", item2, "                           PRICE AFTER DISCOUNT             :   ", final_price2,".rs")
print("ITEM3 : ", item3, "                           PRICE AFTER DISCOUNT             :   ", final_price3,".rs")
print("ITEM4 : ", item4, "                           PRICE AFTER DISCOUNT             :   ", final_price4,".rs")
print("ITEM5 : ", item5, "                           PRICE AFTER DISCOUNT             :   ", final_price5,".rs")
print("")
print("TOTAL PRICE AFTER DISCOUNT                                               :  ", final_price_after_discount,".rs")
print("")
print("GST (18%)                                                                :  ", gst,".rs")
print("")

print("DELIVERY CHARGE                                                          :  ", deliverycharge,".rs")
print("")
print("*****************************************************************************************************************")
print("ACTUAL PRICE AFTER GST, DISCOUNT & DELIVERY CHARGES                      :  ", exact_price,".rs")
print("")
print("*****************************************************************************************************************")
```

```python
#sum of 19 and 20 by while loop
i = 1
sum1 = 0
n1 = 19
while i <= 10:
    pro1 = n1 * i
    print(pro1)
    sum1 = sum1 + pro1
    i = i + 1
print("sum of 19 table:", sum1)
i = 1
sum2 = 0
n2 = 20
while i <= 10:
    pro2 = n2 * i
    print(pro2)
    sum2 = sum2 + pro2
    i = i + 1
print("sum of 20 table:", sum2)

total_sum = sum1 + sum2
avg_19 = sum1 / 10
avg_20 = sum2 / 10
avg_both = (avg_19 + avg_20) / 2

print("total sum of both tables:", total_sum)
print("average of 19 table:", avg_19)
print("average of 20 table:", avg_20)
print("average of both tables:", avg_both)
```

```python
#sum of 19 and 20 by for loop
sum_20t = 0
fact = 1
n1 = 20
for i in range(1, 11):
    fact = n1 * i
    sum_20t = sum_20t + fact
    print(fact)
    print(sum_20t)

sum_19t = 0
fact = 1
n2 = 19
for i in range(1, 11):
    fact = n2 * i
    sum_19t = sum_19t + fact
    print(fact)
    print(sum_19t)

sum_19 = sum_19t
sum_20 = sum_20t
total_sum = sum_19 + sum_20
avg = total_sum / 2

print("sum of table 19:", sum_19)
print("sum of table 20:", sum_20)
print("total sum of tables:", total_sum)
print("avg of table:", avg)
```

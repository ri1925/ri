```python
a = int(input("enter number 1 : "))
b = int(input("enter number 2 : "))

if a > b:
  small = b
else:
  small = a

for i in range(1, small + 1):

  if (a % i == 0) and (b % i == 0):
    hcf = i

print("The HCF of ",a, " and " ,b, " is: ",hcf )
```

```python
a = int(input("enter number 1 : "))
b = int(input("enter number 2 : "))

if a > b:
  greater = a
else:
  greater = b

for i in range(greater, (a * b) + 1):
  if (i % a == 0) and (i % b == 0):
    lcm = i
    break

print("The LCM of", a, "and", b, "is:", lcm)
```

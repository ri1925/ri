```python
print("--------------BILL---------------")

item1 = "Pizza"
price_of_item1 = 500
discount1 = 20

item2 = "Burger"
price_of_item2 = 150
discount2 = 10

item3 = "Fries"
price_of_item3 = 100
discount3 = 15

item4 = "Coffee"
price_of_item4 = 200
discount4 = 30

item5 = "Tea"
price_of_item5 = 50
discount5 = 5



discount_amount1 = price_of_item1 * discount1 / 100
final_price1 = price_of_item1 - discount_amount1

discount_amount2 = price_of_item2 * discount2 / 100
final_price2 = price_of_item2 - discount_amount2

discount_amount3 = price_of_item3 * discount3 / 100
final_price3 = price_of_item3 - discount_amount3

discount_amount4 = price_of_item4 * discount4 / 100
final_price4 = price_of_item4 - discount_amount4

discount_amount5 = price_of_item5 * discount5 / 100
final_price5 = price_of_item5 - discount_amount5


total_bill = final_price1 + final_price2 + final_price3 + final_price4 + final_price5



print("1.st item is   :-          ", item1)
print("")
print("price of it is :-          ", price_of_item1)
print("")
print("discount is    :-          ", discount1, "%")
print("")
print("after discount :-          ", final_price1)
print("")

print("2.nd item is   :-          ", item2)
print("")
print("price of it is :-          ", price_of_item2)
print("")
print("discount is    :-          ", discount2, "%")
print("")
print("after discount :-          ", final_price2)
print("")

print("3.rd item is   :-          ", item3)
print("")
print("price of it is :-          ", price_of_item3)
print("")
print("discount is    :-          ", discount3, "%")
print("")
print("after discount :-          ", final_price3)
print("")

print("4.th item is   :-          ", item4)
print("")
print("price of it is :-          ", price_of_item4)
print("")
print("discount is    :-          ", discount4, "%")
print("")
print("after discount :-          ", final_price4)
print("")

print("5.th item is   :-          ", item5)
print("")
print("price of it is :-          ", price_of_item5)
print("")
print("discount is    :-          ", discount5, "%")
print("")
print("after discount :-          ", final_price5)
print("")

print("Total bill     :-          ", total_bill)
print("")
print("-------------BILL END--------------")
```
